```python
'''
Simple Markov Class

by Daniel Miller - July 2018

'''

import random
from abjad import *

def weighted_random(*weights):
    totalWeight = sum(weights)
    randomInt = random.randint(1, totalWeight)

    for i, weight in enumerate(weights):
        if randomInt <= weight:
            return i
        randomInt -= weight

    return "error"

def get_new_state(switcher, matrix, row):
    return choice(switcher, weighted_random(*matrix[row]))

def choice(switcher, argument):
    return switcher.get(argument, "nothing")


class MarkovChain():
    
    def __init__(self, states, matrix, initialState, upperLimit=10, counter=1):
        self.states = states
        self.matrix = matrix
        self.masterList = [initialState]
        self.upperLimit = upperLimit
        self.counter = counter
        self.switcher = dict(enumerate(states))
    
    def probs(self):
        for index, state in enumerate(self.states):
            print(" ")
            for count in range(0, len(self.states)):
                print(state + " -> " + self.states[count] + " : " + '{0:.2%}'.format(float(self.matrix[index][count]) \
                                                                                /float(sum(self.matrix[index]))))
    def return_chain(self):
        while (self.counter < self.upperLimit):

            for index, state in enumerate(self.states):
                if (self.masterList[-1] == state):
                    self.masterList.append(get_new_state(self.switcher, self.matrix, index))
                    self.counter += 1
            
        print(self.masterList)
```
